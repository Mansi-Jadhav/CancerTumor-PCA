```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\MansiJadhav\\Python-for-Machine-Learning-&-Data-Science-Masterclass\\UNZIP_FOR_NOTEBOOKS_FINAL\\DATA\\cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
help(PCA)
```

```python
pca_model = PCA(n_components=2)
```

```python
pc_results = pca_model.fit_transform(scaled_X)
```

```python
plt.figure(figsize=(8,6), dpi=200)
plt.scatter(pc_results[:,0], pc_results[:,1])
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dictionary = load_breast_cancer()
```

```python
plt.figure(figsize=(8,6), dpi=200)
plt.scatter(pc_results[:,0], pc_results[:,1], c=cancer_dictionary['target'])
```

```python
pca_model.components_    # 2x30
```

```python
df_comp = pd.DataFrame(pca_model.components_, index=['PC1', 'PC2'], columns=df.columns)
```

```python
plt.figure(figsize=(20,3), dpi=150)
sns.heatmap(df_comp, annot=True)
```

```python
pca_model.explained_variance_
```

```python
pca_model.explained_variance_ratio_    # PC1 explains 44% of the data, PC2 explains 19%
```

```python
np.sum(pca_model.explained_variance_ratio_)    # Total 63%
```

```python
pca_30 = PCA(n_components=30)
pca_30.fit(scaled_X)
```

```python
np.sum(pca_30.explained_variance_ratio_)    # 100%
```

```python
explained_variance = []

for n in range(1,30):
    pca_model = PCA(n_components=n)
    pca_model.fit(scaled_X)
    
    explained_variance.append(np.sum(pca_model.explained_variance_ratio_))
```

```python
plt.plot(range(1,30), explained_variance)
plt.xlabel('Number of components')
plt.ylabel('Explained Variance Ratio')
```
